```python
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
```

```python
df.dropna(inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
for col in df.columns:
  if df[col].dtype=='object':
    df[col]=le.fit_transform(df[col])
```

```python
df.info()
```

```python
x=df.drop('Churn',axis=1)
y=df['Churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=23)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=50,random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import *
```

```python
accuracy_score(y_test,y_pred)*100
```

```python
from sklearn.model_selection import GridSearchCV
# GridSearchCV can be applied for all models
```

```python
base_model=RandomForestClassifier(random_state=42)
```

```python
param_grid={
    'n_estimators':[100,200,300],
    'max_depth':[1,5,10],
    'min_samples_split':[2,5,7],
    'min_samples_leaf':[1,2,4],
    'criterion':['gini','entropy'] #splitting techniques, based on error values
}
```

```python
grid_search=GridSearchCV(estimator=base_model,param_grid=param_grid)
```

```python
grid_search.fit(x_train,y_train)
```

```python
print(grid_search.best_params_)
```

```python
final_model=RandomForestClassifier(n_estimators=100,min_samples_split=7,min_samples_leaf=2,max_depth=10,criterion='gini',random_state=42)
```

```python
final_model.fit(x_train,y_train)
```

```python
y_pred_final=final_model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred_final)*100
```
